**Exploratory Data Analysis - Terrorism
The goal of this project is to perform Exploratory Data Analysis (EDA) on the terrorism dataset
to extract meaningful insights and patterns. By analyzing the dataset, we aim to uncover
trends, identify high-risk regions, understand attack characteristics, and potentially reveal
factors that contribute to terrorist activities.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the terrorism dataset
data = pd.read_csv('/content/drive/MyDrive/suspects.csv')

# Display the first few rows to get a sense of the data
print(data.head())

```

```python
# Check for missing values
print(data.isnull().sum())

# Handle missing values (if needed)
# For example, you can drop rows with missing values or impute them with appropriate values.
```

Data Exploration:

```python
# Summary statistics
print(data.describe())

# Data types and missing values
print(data.info())
```

2. Categorical Variables Analysis:

```python
# Countplot for 'sex'
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='sex')
plt.title('Distribution of Suspects by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Countplot for 'marital_status'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='marital_status', order=data['marital_status'].value_counts().index)
plt.title('Distribution of Suspects by Marital Status')
plt.xticks(rotation=45)
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()

# Countplot for 'citizenship_status'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='citizenship_status', order=data['citizenship_status'].value_counts().index)
plt.title('Distribution of Suspects by Citizenship Status')
plt.xticks(rotation=45)
plt.xlabel('Citizenship Status')
plt.ylabel('Count')
plt.show()

```

Analysis of High-Risk Regions:

```python
# Countplot for 'last_residency_state'
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='last_residency_state', order=data['last_residency_state'].value_counts().index)
plt.title('Distribution of Suspects by Last Residency State')
plt.xticks(rotation=90)
plt.xlabel('Last Residency State')
plt.ylabel('Count')
plt.show()

# Countplot for 'last_residency_country'
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='last_residency_country', order=data['last_residency_country'].value_counts().index)
plt.title('Distribution of Suspects by Last Residency Country')
plt.xticks(rotation=90)
plt.xlabel('Last Residency Country')
plt.ylabel('Count')
plt.show()
```

Attack Characteristics:

```python
# Countplot for 'terror_plot'
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='terror_plot')
plt.title('Distribution of Suspects by Terror Plot Involvement')
plt.xlabel('Terror Plot Involvement')
plt.ylabel('Count')
plt.show()

# Countplot for 'military_target'
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='military_target')
plt.title('Distribution of Suspects by Military Target Involvement')
plt.xlabel('Military Target Involvement')
plt.ylabel('Count')
plt.show()

```
